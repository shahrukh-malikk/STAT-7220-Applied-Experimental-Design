Let's say you own a new food truck called **Buster's Breakfast Burritos**. To determine the best ingredients for your burritos, and as a means of minimizing food (and money) waste, you decide to conduct an experiment. Specifically, you want to investigate the following treatments:

| Factor   | Levels                  |
|:---------|:------------------------|
| Protein  | Bacon, Sausage, Chorizo |
| Salsa    | Mild, Hot               |
| Potatoes | Yes, No                 |

To gather data for this experiment, you recruit $r=5$ local residents for each of the $3\times 2\times 2=12$ treatment combinations. After each participant has tasted their randomly assigned burrito, you ask them to rate the burrito on a scale of 1 to 10 (1 = terrible, 10 = amazing). The data are stored in the file `Busters Burritos.xlsx` in the assignment repo. With these data, I want you to:

1. Briefly state the objective of the experiment.




**The objective of the experiment is to determine how the choice of protein, salsa, and the presence of potatoes influence customer's satisfaction with Buster's Breakfast Burritos, so that you can identify the best ingredient combination.**

2. State the outcome variable and how it was measured.

**The outcome variable is customer satisfaction, measured as a (1‚Äì10) rating given by each participant after tasting their assigned burrito.**

3. State the main effects. What lurking variables may be present?

**Main effects:Protein (Bacon, Sausage, Chorizo),Salsa (Mild, Hot) and Potatoes (Yes, No).Possible lurking variables can be Individual taste preferences, hunger level, time of day, spice tolerance, prior experiences with similar foods, and environmental factors such as temperature or presentation.**

4. Why might a full factorial design be preferred here to a $2^k$ fractional factorial design?

**A full factorial design is preferred because all factors have more than two levels (Protein has 3 levels) and the goal is to understand every main effect and interaction without confounding.**

5. State all sets of statistical hypotheses.

Main effects hypotheses:
Protein: H0: All protein types have the same mean satisfaction.
 H1: At least one differs.
Salsa: H0: Mild and Hot have the same mean satisfaction.
 H1: They differ.
Potatoes: H0: Potatoes (Yes/No) have the same mean satisfaction.
 H1: They differ.

Interaction hypotheses:
H0: No interaction between factors.
H1: At least one interaction exists.

Two-Way Interactions

Protein x Salsa:
H0: No interaction between protein and salsa.
H1: Interaction exists.

Protein x Potatoes:
H0: No interaction between protein and potatoes.
H1: Interaction exists.

Salsa x Potatoes:
H0: No interaction between salsa and potatoes.
H1: Interaction exists.

Three-Way Interaction
Protein x Salsa x Potatoes:
H0: No three-way interaction among the three factors.
H1: A three-way interaction exists.

6. Perform appropriate exploratory/descriptive analysis including summary statistics and appropriate visualizations. Do these analyses provide any support for the hypotheses? Briefly explain.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_excel("Busters Burritos.xlsx")

# Convert to category
df["Protein"] = df["Protein"].astype("category")
df["Salsa"] = df["Salsa"].astype("category")
df["Potatoes"] = df["Potatoes"].astype("category")

sns.set(style="whitegrid", context="talk")

# -----------------------
# SUMMARY STATISTICS
# -----------------------

print("===== Overall Summary =====")
print(df["Satisfaction"].describe(), "\n")

print("===== Means by Factor =====")
print(df.groupby("Protein")["Satisfaction"].mean(), "\n")
print(df.groupby("Salsa")["Satisfaction"].mean(), "\n")
print(df.groupby("Potatoes")["Satisfaction"].mean(), "\n")

print("===== Treatment Combinations =====")
print(df.groupby(["Protein", "Salsa", "Potatoes"])["Satisfaction"].mean(), "\n")


# ================================
# MAIN EFFECT: BARPLOTS + BOXPLOTS
# ================================

def main_effect_plots(factor):
    # Barplot
    plt.figure(figsize=(7,5))
    sns.barplot(data=df, x=factor, y="Satisfaction", estimator='mean', ci=95,
                palette="Set2", edgecolor="black")
    plt.title(f"Mean Satisfaction by {factor}")
    plt.ylabel("Mean Satisfaction")
    plt.show()

    # Boxplot
    plt.figure(figsize=(7,5))
    sns.boxplot(data=df, x=factor, y="Satisfaction", palette="Set2")
    plt.title(f"Satisfaction Distribution by {factor}")
    plt.ylabel("Satisfaction")
    plt.show()

main_effect_plots("Protein")
main_effect_plots("Salsa")
main_effect_plots("Potatoes")


# ================================
# INTERACTION PLOTS (Means + Boxplots)
# ================================

def interaction_plots(f1, f2):
    # Interaction Means (Point Plot)
    means = df.groupby([f1, f2])["Satisfaction"].mean().reset_index()

    plt.figure(figsize=(8,5))
    sns.pointplot(data=means, x=f1, y="Satisfaction", hue=f2,
                  markers="o", linestyles="--", palette="Set1")
    plt.title(f"Interaction Plot (Means): {f1} √ó {f2}")
    plt.ylabel("Mean Satisfaction")
    plt.show()

    # Interaction Boxplot
    plt.figure(figsize=(8,5))
    sns.boxplot(data=df, x=f1, y="Satisfaction", hue=f2, palette="Set2")
    plt.title(f"Interaction Boxplot: {f1} √ó {f2}")
    plt.ylabel("Satisfaction")
    plt.show()

interaction_plots("Protein", "Salsa")
interaction_plots("Protein", "Potatoes")
interaction_plots("Salsa", "Potatoes")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel("Busters Burritos.xlsx")

# Convert to categories
df["Protein"]   = df["Protein"].astype("category")
df["Salsa"]     = df["Salsa"].astype("category")
df["Potatoes"]  = df["Potatoes"].astype("category")

sns.set(style="whitegrid", context="talk")

# -----------------------
# SUMMARY STATISTICS
# -----------------------

print("===== Overall Summary =====")
print(df["Satisfaction"].describe(), "\n")

print("===== Means by Factor =====")
print(df.groupby("Protein")["Satisfaction"].mean(), "\n")
print(df.groupby("Salsa")["Satisfaction"].mean(), "\n")
print(df.groupby("Potatoes")["Satisfaction"].mean(), "\n")

print("===== Treatment Combinations =====")
print(df.groupby(["Protein", "Salsa", "Potatoes"])["Satisfaction"].mean(), "\n")


# ================================
# MAIN EFFECT BARPLOTS (simple)
# ================================

def barplot_main(factor):
    plt.figure(figsize=(7,5))
    sns.barplot(
        data=df, x=factor, y="Satisfaction",
        estimator="mean", ci=95, palette="Set2", edgecolor="black"
    )
    plt.title(f"Mean Satisfaction by {factor}")
    plt.ylabel("Mean Satisfaction")
    plt.show()

barplot_main("Protein")
barplot_main("Salsa")
barplot_main("Potatoes")


# ================================
# INTERACTION PLOTS (BOXPLOT + MEAN LINE)
# ================================

def interaction_boxplot_mean(f1, f2):
    # BOX PLOT WITH HUE (shows interaction & spread)
    plt.figure(figsize=(8,5))
    sns.boxplot(
        data=df, x=f1, y="Satisfaction", hue=f2,
        palette="Set2"
    )
    plt.title(f"Interaction Boxplot: {f1} √ó {f2}")
    plt.ylabel("Satisfaction")
    plt.show()

    # MEAN LINE PLOT (cleanest interaction display)
    means = df.groupby([f1, f2])["Satisfaction"].mean().reset_index()

    plt.figure(figsize=(8,5))
    sns.pointplot(
        data=means, x=f1, y="Satisfaction", hue=f2,
        markers="o", linestyles="--", palette="Set1"
    )
    plt.title(f"Interaction Means: {f1} √ó {f2}")
    plt.ylabel("Mean Satisfaction")
    plt.show()


# Apply to all 3 two-way interactions
interaction_boxplot_mean("Protein", "Salsa")
interaction_boxplot_mean("Protein", "Potatoes")
interaction_boxplot_mean("Salsa", "Potatoes")
```
