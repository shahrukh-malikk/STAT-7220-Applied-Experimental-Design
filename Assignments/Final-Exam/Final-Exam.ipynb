```python
## Question 6 Code ##

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_excel("Busters Burritos.xlsx")

# Convert to categorical
df["Protein"]  = df["Protein"].astype("category")
df["Salsa"]    = df["Salsa"].astype("category")
df["Potatoes"] = df["Potatoes"].astype("category")

sns.set(style="whitegrid", context="talk")


# MAIN-EFFECT
def barplot_main(factor):
    plt.figure(figsize=(7, 5))
    sns.barplot(
        data=df,
        x=factor,
        y="Satisfaction",
        estimator="mean",
        ci=95,
        palette="Set2",
        edgecolor="black"
    )
    plt.title(f"Mean Satisfaction by {factor}")
    plt.ylabel("Mean Satisfaction")
    plt.xlabel(factor)
    plt.tight_layout()
    plt.show()

barplot_main("Protein")
barplot_main("Salsa")
barplot_main("Potatoes")

# INTERACTION: BOX + MEAN LINE
def interaction_box_line(f1, f2, title=None):
    if title is None:
        title = f"Interaction Plot: {f1} √ó {f2}"

    fig, ax = plt.subplots(figsize=(10, 7))

    # Boxplot shows distribution within each combination
    sns.boxplot(
        data=df,
        x=f1,
        y="Satisfaction",
        hue=f2,
        dodge=True,
        ax=ax,
        zorder=1
    )

    # Overlaid mean line with shaded CI (same x & hue)
    sns.lineplot(
        data=df,
        x=f1,
        y="Satisfaction",
        hue=f2,
        estimator="mean",
        ci=95,
        marker="o",
        linestyle="--",
        ax=ax,
        zorder=2
    )

    # Remove duplicate legend entries (from box + line)
    handles, labels = ax.get_legend_handles_labels()
    by_label = dict(zip(labels, handles))   # unique labels
    ax.legend(by_label.values(), by_label.keys(), title=f2)

    ax.set_title(title, fontsize=16)
    ax.set_ylabel("Satisfaction")
    ax.set_xlabel(f1)
    plt.tight_layout()
    plt.show()

# Two-way interactions (with your desired style)
interaction_box_line("Protein", "Salsa",    "Interaction Plot: Protein √ó Salsa")
interaction_box_line("Protein", "Potatoes", "Interaction Plot: Protein √ó Potatoes")
interaction_box_line("Salsa",   "Potatoes", "Interaction Plot: Salsa √ó Potatoes")
```

```python

# 3-WAY INTERACTION:
# Protein √ó Salsa √ó Potatoes
# (boxplots + mean lines, faceted by Salsa)

def three_way_interaction_box_line():
    # Boxplots in separate panels for each Salsa level
    g = sns.catplot(
        data=df,
        x="Protein",
        y="Satisfaction",
        hue="Potatoes",
        col="Salsa",
        kind="box",
        dodge=True,
        height=5,
        aspect=1.0
    )

    g.fig.subplots_adjust(top=0.82)
    g.fig.suptitle("Three-Way Interaction: Protein √ó Salsa √ó Potatoes", fontsize=16)

    # Compute means for each Protein √ó Potatoes √ó Salsa combo
    means = (
        df.groupby(["Salsa", "Protein", "Potatoes"])["Satisfaction"]
          .mean()
          .reset_index()
    )

    # Overlay mean lines in each facet (one panel per Salsa)
    salsa_levels = sorted(df["Salsa"].unique())
    for ax, salsa in zip(g.axes.flat, salsa_levels):
        sub = means[means["Salsa"] == salsa]
        sns.pointplot(
            data=sub,
            x="Protein",
            y="Satisfaction",
            hue="Potatoes",
            markers="o",
            linestyles="--",
            dodge=0.3,
            ax=ax,
            errorbar=None
        )
        ax.set_title(f"Salsa: {salsa}")
        ax.set_xlabel("Protein")
        ax.set_ylabel("Satisfaction")

    # Fix legend: keep only one, outside
    handles, labels = g.axes.flat[0].get_legend_handles_labels()
    # First legend from boxplot, second from pointplot ‚Äì keep unique
    by_label = dict(zip(labels, handles))
    g.fig.legend(
        by_label.values(),
        by_label.keys(),
        title="Potatoes",
        loc="upper right",
        bbox_to_anchor=(0.98, 0.92)
    )

    # Remove per-axes legends
    for ax in g.axes.flat:
        ax.legend_.remove()

    plt.show()

# Call it
three_way_interaction_box_line()
```

7. Fit the full factorial ANOVA model. Check the assumption of normality using a testing method and a visual method. Does the assumption hold? Why or why not?

```python
## Question 7 Code ##

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro


df["Protein"]  = df["Protein"].astype("category")
df["Salsa"]    = df["Salsa"].astype("category")
df["Potatoes"] = df["Potatoes"].astype("category")

#  ANOVA model
full_model = ols('Satisfaction ~ C(Protein) * C(Salsa) * C(Potatoes)', data=df).fit()
# Using type 2 or type 3 anova for factorial designs with interactions
print(sm.stats.anova_lm(full_model, typ=2))

# --- Normality Assumption Check ---
# Q-Q plot of residuals
residuals = full_model.resid

plt.figure(figsize=(8, 6))
# Q-Q plot of the residuals
sm.qqplot(residuals, line='s')
plt.title('Q-Q Plot of Residuals')
plt.show()

sns.histplot(full_model.resid, kde=True)
plt.title("Histogram of Residuals")
plt.show()

# Shapiro-Wilk Test
shapiro_test = shapiro(residuals)
print(f"\nShapiro-Wilk Test for Normality:")
print(f"Statistic: {shapiro_test.statistic:.3f}")
print(f"p-value: {shapiro_test.pvalue:.3f}")

# Interpret the Shapiro-Wilk test
alpha = 0.05
if shapiro_test.pvalue > alpha:
    print(f"(p-value > {alpha}) => Fail to reject H0: Residuals appear normally distributed.")
else:
    print(f"(p-value <= {alpha}) => Reject H0: Residuals do not appear normally distributed.")
```

The ANOVA results show significant main effects for Protein and Salsa. Potatoes is not statistically significant. None of the interaction terms are significant, which suggests that each ingredient‚Äôs effect on satisfaction operates mostly independently rather than depending on other ingredients.

For the normality assumption, the Q‚ÄìQ plot shows minor tail deviations, but the Shapiro‚ÄìWilk test is not significant (p = 0.135). This means the residuals do not significantly depart from normality. Overall, the normality assumption is reasonably satisfied.

8. Check the assumption of constant variance using a testing method and a visual method. Does the assumption hold? Why or why not?

```python
## Question 8 Code ##

from statsmodels.stats.diagnostic import het_breuschpagan
from scipy.stats import levene
import matplotlib.pyplot as plt
import seaborn as sns


fitted_values = full_model.fittedvalues
residuals = full_model.resid

plt.figure(figsize=(10,8))
sns.scatterplot(x=fitted_values, y=residuals, color='black')
plt.axhline(y=0, color='red', linestyle='--')
plt.title("Residuals vs Fitted Values")
plt.axhline(y=3, color='blue')
plt.axhline(y=-3, color='blue')
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.tight_layout()
plt.show()


# Levene‚Äôs Test for Homogeneity of Variance

groups = [group["Satisfaction"].values
          for _, group in df.groupby(["Protein", "Salsa", "Potatoes"])]

levene_stat, levene_p = levene(*groups)

print("Levene‚Äôs Test for Homogeneity of Variance:")
print(f"Statistic = {levene_stat:.4f}, p-value = {levene_p:.4f}")

bp_stat, bp_pvalue, f_stat, f_pvalue = het_breuschpagan(full_model.resid, full_model.model.exog)

print("\nBreusch‚ÄìPagan Test for Homogeneity of Variance:")
print(f"BP Statistic = {bp_stat:.4f}, BP p-value = {bp_pvalue:.4f}")
print(f"F-statistic = {f_stat:.4f}, F p-value = {f_pvalue:.4f}")
```

Residuals vs fitted plot: looks random , supports constant variance
Levene‚Äôs test: p > 0.05 : supports constant variance
Breusch‚ÄìPagan test: p < 0.05 : suggests variance is not constant

Two methods support the constant variance assumption, but one method (Breusch‚ÄìPagan) does not.hence we reject  the Null.

7. Assuming a defining relation of $I=ABC$, specify the aliasing structure of the design. What does this mean for the interpretation of the main effects and interactions?

8. Fit the appropriate ANOVA model to the data. Assess the normality assumption using two methods. Do we have evidence to support the assumption of normality? Why or why not?

**Answer 7:** Main Effects:

A = BC
B = AC
C = AB

Two-Factor Interactions:

AB = C
AC = B
BC = A

Three-Factor Interaction:
ABC = I

Interpretation:
Because each main effect is aliased with a two-factor interaction, we cannot interpret any main effect independently. An observed effect for A, for example, actually represents the combined influence of A + BC, and the same applies to B and C. This means the design does not allow us to separate main effects from the interactions they are confounded with.

```python
## Question 8 Code ##

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import shapiro
import seaborn as sns
import matplotlib.pyplot as plt


print("\n=== ANOVA: Sales ~ Location + Timing + Marketing ===")

model = ols("Sales ~ Location + Timing + Marketing", data=burritos).fit()
anova_table = anova_lm(model, typ=2)
print("\nANOVA Table:")
print(anova_table)

# NORMALITY ASSUMPTION

# 1. Shapiro-Wilk Test
w, p_value = shapiro(model.resid)
print("\nShapiro‚ÄìWilk Test:")
print("W =", w, "| p-value =", p_value)

# 2. QQ Plot
sm.qqplot(model.resid, line='s')
plt.title("QQ Plot of Residuals")
plt.show()

# Histogram
sns.histplot(model.resid, kde=True)
plt.title("Residuals Histogram")
plt.show()
```

```python
## Question 9 Code ##

from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan
import scipy.stats as stats
import matplotlib.pyplot as plt


# Standardized residuals and fitted values

influence = model.get_influence()
std_resid = influence.resid_studentized_internal
fitted = model.fittedvalues


# Residuals vs Fitted Plot

plt.figure(figsize=(8, 6))
plt.scatter(fitted, std_resid)
plt.axhline(0, color='red', linestyle='--')
plt.axhline(3, color='blue')
plt.axhline(-3, color='blue')
plt.xlabel("Fitted Values")
plt.ylabel("Standardized Residuals")
plt.title("Residuals vs Fitted")
plt.show()

# Breusch‚ÄìPagan Test
bp_stat, bp_pvalue, f_stat, f_pvalue = het_breuschpagan(
    model.resid,
    model.model.exog
)

print("\nBreusch‚ÄìPagan Test:")
print(f"BP Statistic = {bp_stat:.4f}, BP p-value = {bp_pvalue:.4f}")
print(f"F-statistic = {f_stat:.44f}, F p-value = {f_pvalue:.4f}")


# Levene‚Äôs Test
groups = [
    g["Sales"].values
    for _, g in burritos.groupby(["Location", "Timing", "Marketing"])
]

lev_stat, lev_p = stats.levene(*groups)

print("\nLevene‚Äôs Test for Homogeneity of Variance:")
print(f"Statistic: {lev_stat}")
print(f"p-value:   {lev_p}")

```

The residuals-versus-fitted plot shows the points scattered fairly evenly around zero with no apparent funnel shape, suggesting that the variance appears roughly constant based on visual inspection. The Breusch‚ÄìPagan test gives a p-value of 0.0662, which is slightly above 0.05, so we fail to reject the null hypothesis of constant variance; this test does not provide strong evidence of heteroscedasticity. However, Levene's test tells a very different story, with an extremely large test statistic and an extremely small p-value (p ‚âà 2√ó10‚Åª¬π¬π‚Å∏), indicating overwhelming evidence that the group variances are not equal. Together, these results suggest that although the plot and the BP test imply approximate homoscedasticity, the more sensitive Levene's test strongly indicates that the constant variance assumption is violated.
